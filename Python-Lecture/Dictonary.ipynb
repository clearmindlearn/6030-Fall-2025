```python
# Dictionary
#0 How to create a dict
#1. Mutable or Immutable
#2. Access element
#3. Add
#4 Remove
```

```python
#Create a dict
john = {'name': 'John', 'age':45, 'country':'USA'} #key:value pair
print (john)

kevin = dict(name='Kevin', age=26, country='UK')
print(kevin)
```

```python
fruits = {"apple":1.45, "banana":2.40, "cherry":5.45 }
print (fruits)
```

```python
# Mutable or Immutable
john['age'] = 50
print(john)
#Dict ---> Mutable!!!!!!!!!

john['country'] = "UK"
print(john)
```

```python
#Acess
print(john['name'])
print(john['country'])
```

```python
# ADD
john['height'] = 6.00
print(john)
```

```python
#remove method 1 --> pop
john.pop('height')
print(john)
```

```python
#Remove method 2 --> del
del john['age']
print(john)
```
