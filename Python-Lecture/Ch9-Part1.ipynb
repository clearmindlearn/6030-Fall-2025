```python
# Class defition = Attributes + Methods()

class Student:  #Let's define the class Student!
   #Main goal of the Constuctor --> introduce (and initialize ) the attributes
   #Define 4 attributes: name, age, major and id

   def __init__(self,n,a,maj,i): # __init__ --> constructor of the class Student
      self.name = n  #introduce name attribute and initialize name as n
      self.age = a
      self.major = maj #major -> attribute, maj -> parameter of function __init__
      self.id = i


#Let's create methods() for the class Student!
#OOP --> set & get methods

   def get_fullname(self): # just print the name of the student
     print("Full Name is :", self.name)
   def get_age(self):
     print("Age is :", self.age)

   def set_name(self, new_name):
      self.name = new_name
```
