```python
# Type 1
score = 98

if score >= 90:
  print("your letter grade is A")
  print("Congratulations!")

```

```python
# Type 2 --> if else

#if statement --> logical /boolean expression --> evaluated to be either True or False!

score = 88

if score >= 90:
  print("your letter grade is A")
else:
  print("your letter grade is NOT A ")
```

```python
#Type 3 --> if elseif ... else

score = int(input("Please enter your score"))

if score >= 90:
  print("your letter grade is A")
elif score >= 80:
  print("your letter grade is B")
elif score >= 70:
  print("your letter grade is C")
else:
  print("Fail")
```

```python
# Relational operator --> boolean/logical expression --> evaluated to be True or False
x = (3 > 4)
print (x)

y = (3 == 4)
print (y)

z = (3 != 4)
print (z)

m = (3 <= 4)
print(m)
```

```python
#Logical  operator  --> AND  OR  and NOT (logical table) --> evaluated to be True or False

x = 7
y = 9

print((x > 0) and (y < 10))
print((x > 10)  or (y < 11))
print(not(x > 10))
```
