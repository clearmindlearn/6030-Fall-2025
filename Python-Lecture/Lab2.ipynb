 Use % to get the desired rightmost digits. Ex: The rightmost 2 digits of 572 is gotten by 572 % 100, which is 72.



```python
572%10  #keep the one righmost digit
```

```python
572%100 #keep the two righmost digits
```

Use // to shift right by the desired amount. Ex: Shifting 572 right by 2 digits is done by 572 // 100, which yields 5. (Recall integer division discards the fraction).

```python
572 // 10 #drop the one rightmost digit --> I just drop 2
```

```python
572 // 100 #drop the two rightmost digits --> I just drop 72
```
