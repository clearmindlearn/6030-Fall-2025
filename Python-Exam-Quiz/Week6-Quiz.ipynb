```python
def grade_score(score):
    if score >= 60:
        if score >= 80:
            if score >= 90:
                return "A"
            else:
                return "B"
        else:
            return "C"
    else:
        return "F"
grade_score(85)
```

```python
def classify(x, y):
    if x > 0:
        if y > 0:
            if x > y:
                return "x"
            else:
                return "y"
        else:
            return "yy"
    else:
        if y > 0:
            return "xp"
        else:
            return "bp"


print(classify(3, 5))
```
