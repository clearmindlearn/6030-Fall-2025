```python
# count the number of below-freezing Celsius temperatures in the list

#scanning a list (array)  -> better to use for-loop rather than while-loop

# for-loop --> element approach

temperatures = [30, 20, 2, -5, -15] #list

num_neg = 0

for temp in temperatures:
    if temp < 0:
      num_neg = num_neg + 1


print("answer is", num_neg)
```

```python
#Another way to write a for-loop --> index approach
temperatures = [30, 20, 2, -5, -15] #list

num_neg = 0

for index in range(len(temperatures)):
    if temperatures[index] < 0:
      num_neg = num_neg + 1


print("answer is", num_neg)
```

```python
mid_score = [100, 87, 76, 81, 92]

#index approach

num_B = 0

for i in range(len(mid_score)):
  if (mid_score[i]  <90) & (mid_score[i] >= 80):
    num_B = num_B + 1

print(num_B)  # find # of students receiving B

```
