```python
#List
# 1. Mutable Array
# 2. operations (not an element wise operation!!)
# 3. add, remove
```

```python
# Mutable Array!!!
# Index starts 0
fruits = ["green grape", "blueberry", "cherry"]
print(fruits[0])
print(fruits[1])
print(fruits[2])
print(len(fruits))

# Mutable!!!!!!!!!!!!!!! (not lik string)
fruits[0] = "apple"
print(fruits)

```

```python
# List operations

midterm_scores = [100, 98, 78,82, 88] #I have 5 students; their midterm scores
print(midterm_scores)
finalexam_scores = [85,95,90, 95, 100]
print(finalexam_scores)

#midterm score + finalexam_score
total_score = midterm_scores + finalexam_scores # take a look at the result below! Is what you expect?

print(total_score)
```

```python
#inserting an element to list

fruits = ["green grape", "blueberry", "cherry"]
fruits.insert(1, "apple") # add apple at index 1
print(fruits)
```

```python
fruits = ["green grape", "blueberry", "cherry"]
fruits.append("apple") # add apple at the end
print(fruits)
```

```python
# Remove the item from Fruits
fruits = ["green grape", "blueberry", "cherry"]
fruits.remove("blueberry")
print(fruits)
```

```python
# Remove the element from fruits using pop()
fruits = ["green grape", "blueberry", "cherry"]
fruits.pop(1)
print(fruits)
```
