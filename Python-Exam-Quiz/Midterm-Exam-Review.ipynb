```python
# count # of even numbers
num = [1,3,4,5,6,8]
count = 0
#use the for loop
for i in range(len(num)):
  if num[i] %2 == 0:
    count = count + 1  # even

print (count)
```

```python
count = 0
for element in num:
  if element %2 == 0:
    count = count + 1  # even

print (count)
```
